```python
import matplotlib.pyplot as plt
```

```python

x = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y = [1400, 2500, 5800, 3200, 7600, 60000, 85000]
plt.title("Highest Sale on Which Day")
plt.plot(x, y, color='black', linestyle='--', linewidth=2, marker='o', label='Daily Sales')
plt.xlabel("Week Days")
plt.ylabel("Highest Sale")
plt.legend()
plt.show()
```

```python
x = [2014,2016,2018,2022,2024,2025]
y = [220000,170000,330000,44000,74000,800000]
plt.xlabel("Year")
plt.ylabel("Sales Data")
plt.title("Sales analysis")
plt.plot(x,y,color="Red",marker="*")
plt.grid()
plt.show()
```
